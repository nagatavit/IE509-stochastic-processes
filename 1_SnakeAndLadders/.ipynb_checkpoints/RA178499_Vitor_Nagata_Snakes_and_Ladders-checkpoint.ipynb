## Snakes and Ladders

Inicializando os valores e bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
MAX_ROUNDS = 200

transition_matrix = np.zeros((101,101))
prob_on_current_turn = np.zeros(101)
prob_on_next_turn = np.zeros(101)

# dict to convert positions where there is snakes and ladders
snakes_and_ladders = {
    # ladders
    1 : 38, 
    4 : 14,
    9 : 31,
    21 : 42,
    28 : 84,
    36 : 44,
    51 : 67,
    71 : 91,
    80 : 100,
    # snakes
    98 : 78,
    95 : 75,
    93 : 73,
    87 : 24,
    64 : 60,
    62 : 19,
    56 : 53,
    49 : 11,
    48 : 26,
    16 : 6
}
```

### Matriz de Transição

```python
# This function checks the possible outcomes for each of
# the 6 possible dice rolls in a given start position
def possible_outcomes(start):
    # positions if there wasn't snakes or ladders
    outcomes = np.arange(start+1, start+7)
    
    # substituting snakes and ladders positions
    for i in range(len(outcomes)):
        
        # start position is not at the bottom of a ladder or the head of
        # a snake
        if start not in snakes_and_ladders:
            outcomes[i] = snakes_and_ladders.get(outcomes[i], outcomes[i])
            
            # positons over 100 are also considered a win
            if (outcomes[i] > 100):
                outcomes[i] =  100
                
        # if the start position was the bottom of a ladder or the head of a
        # snake, the possible outcomes is zero.
        else:
            outcomes[i] = 0
    
    return(outcomes)
```

```python
# transition matrix
for idx in range(101):
    
    # checks to see if the starting position is possible (isn't on a snake or ladder)
    if idx not in snakes_and_ladders:
        next_roll_outcomes = possible_outcomes(idx)
    
        # for each of the outcomes, add the probability of 1/6
        for position in next_roll_outcomes:
            transition_matrix[position, idx] = transition_matrix[position, idx] + 1/6
        
```

### Iterações das rodadas

```python
# initialize the probability on the first turn to 
# be on the position 0
prob_on_current_turn[0] = 1
prob_on_next_turn[0] = 1

# probabilities of winning and to have already won
current_prob_of_winning = np.zeros(MAX_ROUNDS)
cumulative_prob_of_winning = np.zeros(MAX_ROUNDS)

# iterations for each turn
for turn in range(MAX_ROUNDS):
    # stores the probability to have already won in the current turn
    cumulative_prob_of_winning[turn] = prob_on_current_turn[-1]
    
    # calculates the next probabilities
    prob_on_next_turn = transition_matrix.dot(prob_on_current_turn)
    
    # stores the probability of winning exactly on this turn
    current_prob_of_winning[turn] = prob_on_next_turn[-1] - prob_on_current_turn[-1]
    
    # prepare for the next iteration
    prob_on_current_turn = prob_on_next_turn.copy()
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.ylabel("chance of winning")
plt.xlabel("turns")
plt.plot(current_prob_of_winning)
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.ylabel("chance to have already won")
plt.xlabel("turns")
plt.plot(cumulative_prob_of_winning)
```
